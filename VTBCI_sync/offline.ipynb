```python
import os
import pickle

cwd = os.getcwd()
fatherpath = 'OperationSystem\ResultStored'
personName = 'diode'

with open(os.path.join(cwd, fatherpath, personName,'test/raw.pickle'),'rb') as fp:
    data = pickle.load(fp)
with open(os.path.join(cwd, fatherpath, personName,'info/info.pickle'),'rb') as fp:
    info = pickle.load(fp)
```

```python
import math
import numpy as np

trackTimes = info['trackTimes']
tktgs = info['TK_TG']
tktgs = tktgs[:len(trackTimes)]
Nx, Ny = info['laout']
# Nx, Ny = info['layout']
Wx, Wy = info['real_wz']

Fitts = []
TRD = []
for i in range(len(trackTimes)):
    distance = tktgs[i].real
    time = trackTimes[i]
    Fitts.append(math.log2(distance*2)/time)
    TRD.append(math.log2(math.sqrt(Nx*Ny))/math.sqrt(Wx*Wy)*distance/time*100)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(5,5), dpi=300)
plt.scatter(np.arange(len(Fitts))+1, Fitts, color='k')
xlim = ax.get_xlim()
plt.hlines(np.mean(Fitts), xlim[0], xlim[1], colors='k', linestyles='--')
plt.xlabel('epoch')
plt.ylabel('Fitts ITR')
```

```python
fig, ax = plt.subplots(figsize=(5,5), dpi=300)
plt.scatter(np.arange(len(TRD))+1, TRD, color='k')
xlim = ax.get_xlim()
plt.hlines(np.mean(TRD), xlim[0], xlim[1], colors='k', linestyles='--')
plt.xlabel('epoch')
plt.ylabel('TRD')
```
